# Análisis Exploratorio de Datos

1. Carga de datos

```python
import pandas as pd

data_train = pd.read_csv('data/train.csv')
data_test = pd.read_csv('data/test.csv')
```

2. Exploración del dataset

```python
print(data_train.head()) # Exploración de las primeras filas para comprender los datos
print(data_train.info()) # Tipo de datos
print(data_train.describe()) # Valores estadísticos del dataset
```

3. Manejo de datos faltantes

```python
# Visualización de valores faltantes
import seaborn as sns
sns.heatmap(data_train.isnull(), cbar=False)
```

```python
# Inputación de valores en columnas con valores faltantes
data_train['Age'].fillna(data_train['Age'].median(), inplace=True) # Sustituimos valores NaN por el valor de la mediana
data_train['Embarked'].fillna(data_train['Embarked'].mode()[0], inplace=True)
data_train.drop('Cabin', axis=1, inplace=True) # Eliminamos esta columna porque tiene demasiados valores NaN
```

# Visualización de Datos

```python
# Gráfica de supervivencia por género
import seaborn as sns
sns.countplot(data=data_train, x='Sex', hue='Survived')

# Gráfica de supervivencia por clase y edad
sns.violinplot(data=data_train, x='Pclass', y='Age', hue='Survived', split=True)
```
